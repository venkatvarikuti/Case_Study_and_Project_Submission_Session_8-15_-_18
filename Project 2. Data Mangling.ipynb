```python
17. Build a dataset with every possible (UserID, ProductID) pair (cross join)
```

```python
df_userrid = pd.DataFrame({'UserID': users['UserID']})
df_prodId = pd.DataFrame({'ProductID': products['ProductID']})

#create new column Key with value as 1 for both the dataframe as this would become the common key to be merged
df_userrid['Key'] = 1
df_prodId['Key'] = 1

df_out = pd.merge(how='outer',left=df_userrid,right=df_prodId,on='Key')[['UserID','ProductID']]
df_out.head(13)
```

```python
18. Determine how much quantity of each product was purchased by each user
```

```python
df_product = pd.merge(how='left',left=df_out,right=transactions,on=['UserID','ProductID'])
df_user_quantity= df_product.groupby(by=['UserID','ProductID'])['Quantity'].sum()
#reset index so that the index column will have consecutive default numbers and fill NAN values with 0
df_user_quantity.reset_index().fillna(0)
```

```python
#Joinn left outer join users with transactions
df_user_trans = pd.merge(how='left',left=users,right=transactions,on='UserID')
df_user_trans = df_user_trans.drop_duplicates(subset='UserID')
df_user_trans.reset_index(drop=True)
```

```python
 missing_info = list(df_user_trans.columns[df_user_trans.isnull().any()])
missing_info
```

```python
for col in missing_info:
    num_missing = df_user_trans[df_user_trans[col].isnull() == True].shape[0] 
    print('numbermissing for column {}: {}'.format(col, num_missing))
```

```python
#count of missing data
for col in missing_info:
    percent_missing = df_user_trans[df_user_trans[col].isnull() == True].shape[0] / df_user_trans.shape[0] 
    print('percent missing for column {}: {}'.format( col,percent_missing))
```
